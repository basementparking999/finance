```python
#pdf 를 text 로 변환하기
import os
from os import listdir
from os.path import isfile, join

import tika
from tika import parser

def pdf2txt (source_folder='./pdf/', output_folder="./text/") :
    # 지정 폴더 내 파일 목록 조회 (파일만)
    pdf_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    
    for pdf in pdf_files :
        pdf_filepath = source_folder + pdf
        # pdf 파일을 text로 변환
        print(pdf_filepath)
        parsedPDF = parser.from_file(pdf_filepath)["content"]
        
        with open((output_folder + pdf).replace('.pdf', '.txt'), 'w', encoding='UTF8') as f :
            f.write(parsedPDF)
            f.close()    
            
pdf2txt()
```
